### 1. Data Preprocessing on flight data

```python
data_2020_01 = pd.read_csv("data/flightlist_20200101_20200131.csv")
data_2020_02 = pd.read_csv("data/flightlist_20200201_20200229.csv")
data_2020_03 = pd.read_csv("data/flightlist_20200301_20200331.csv")
data_2020_04 = pd.read_csv("data/flightlist_20200401_20200430.csv")
data_2020_05 = pd.read_csv("data/flightlist_20200501_20200531.csv")
data_2020_06 = pd.read_csv("data/flightlist_20200601_20200630.csv")
data_2020_07 = pd.read_csv("data/flightlist_20200701_20200731.csv")
data_2020_08 = pd.read_csv("data/flightlist_20200801_20200831.csv")
data_2020_09 = pd.read_csv("data/flightlist_20200901_20200930.csv")                           
data_2020_10 = pd.read_csv("data/flightlist_20201001_20201031.csv")
data_2020_11 = pd.read_csv("data/flightlist_20201101_20201130.csv")
data_2020_12 = pd.read_csv("data/flightlist_20201201_20201231.csv")                           
```

```python
frames = [data_2020_01, data_2020_02, data_2020_03, data_2020_04, data_2020_05, data_2020_06, data_2020_07, data_2020_08, data_2020_09, data_2020_10, data_2020_11, data_2020_12]

data_2020_01_12 = pd.concat(frames)

print(data_2020_01_12.shape)
```

```python
data_2020_01_12.head()
```

```python
data_2020_01_12_raw = data_2020_01_12[["number","typecode", "origin", "destination", "day"]].reset_index(drop=True)
```

```python
data_2020_01_06_raw.to_csv('data/data_2020_01_06.original.csv', index=True)
```

```python
data_2020_01_12_filtered = data_2020_01_12.dropna(subset=['origin', 'destination'])
data_2020_01_12_filtered = data_2020_01_12_filtered.reset_index(drop=True)
data_2020_01_12_filtered.shape
```

```python
data_2020_01_12_filtered = data_2020_01_12_filtered[["number","typecode", "origin", "destination", "day"]]
```

```python
data_2020_01_06_filtered.to_csv('data/data_2020_01_06_filtered.csv', index=True)
```

### 2. Combine flight data with airport data

```python
airports = pd.read_csv('data/airports.csv')
airport_frequencies = pd.read_csv('data/airport-frequencies.csv')
regions = pd.read_csv('data/regions.csv')
```

```python
print(airports.shape)
print(airport_frequencies.shape)
print(regions.shape)
```

```python
airports.head()
```

```python
airports = airports[["ident","type", "name", "latitude_deg", 
                     "longitude_deg", "continent", "iso_country", "iso_region", "municipality" ]]
```

```python
airports.head()
```

```python
airports.to_csv('data/airports.csv', index=False)
```

### 3. Merge flight info & airports 

```python
origin_airports = airports
destination_airports = airports
```

```python
origin_airports.columns = ["origin_ident","origin_type", "origin_name", "origin_latitude_deg", 
                     "origin_longitude_deg", "origin_continent", "origin_iso_country", "origin_iso_region", "origin_municipality" ]
```

```python
results = pd.merge(data_2020_01_12_filtered, origin_airports, how="left", left_on = "origin", right_on = "origin_ident")
results
```

```python
destination_airports.columns = ["destination_ident","destination_type", "destination_name", "destination_latitude_deg", 
                                "destination_longitude_deg", "destination_continent", "destination_iso_country", "destination_iso_region", 
                                "destination_municipality" ]

```

```python
results = pd.merge(results, destination_airports, how="left", left_on = "destination", right_on = "destination_ident")
results
```

```python
results.shape
```

```python
international_flight_2020 = results[results["origin_iso_country"] != results["destination_iso_country"]]
```

```python
international_flight_2020.shape
```

```python
national_flight_2020 = results[results["origin_iso_country"] == results["destination_iso_country"]]
```

```python
national_flight_2020.shape
```

```python
international_flight_2020.to_csv("data/international_flight_all_2020.csv", index=False)
national_flight_2020.to_csv("data/national_flight_all_2020.csv", index=False)
```
