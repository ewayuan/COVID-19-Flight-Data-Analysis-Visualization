```python
def build_graph(df, airports_info):
    directed_G = nx.DiGraph()
    directed_G.add_nodes_from(airports_info)
    for index, row in df.iterrows():
        if directed_G.has_edge(row['origin'], row['destination']):
                directed_G.edges[row['origin'], row['destination']]['weight'] += 1
                directed_G.edges[row['origin'], row['destination']]['date'].add(row['date'])
        else:
            directed_G.add_edge(row['origin'], row['destination'], date=set(),weight=1)
            directed_G.edges[row['origin'], row['destination']]['date'].add(row['date'])
    return directed_G
```

```python
def build_daily_graph(df, days, airports_info):
    graph_list = []
    for day in days:
        cur_day_flights = df[df['day'] == day]
        directed_G = nx.DiGraph()
        directed_G.add_nodes_from(airports_2019_01_01)
#         for airport in airports:
#             directed_G.add_no de(airport)
        for index, row in cur_day_flights.iterrows():
            if directed_G.has_edge(row['origin'], row['destination']):
                directed_G.edges[row['origin'], row['destination']]['weight'] += 1
            else:
                directed_G.add_edge(row['origin'], row['destination'], weight=1)
        graph_list.append(directed_G)
    return graph_list
```
