```python
import pandas as pd
import numpy as np
```

### 1. Data Preprocessing on flight data

```python
data_2019_01 = pd.read_csv("data/flightlist_20190101_20190131.csv")
data_2019_02 = pd.read_csv("data/flightlist_20190201_20190228.csv")
data_2019_03 = pd.read_csv("data/flightlist_20190301_20190331.csv")
data_2019_04 = pd.read_csv("data/flightlist_20190401_20190430.csv")
data_2019_05 = pd.read_csv("data/flightlist_20190501_20190531.csv")
data_2019_06 = pd.read_csv("data/flightlist_20190601_20190630.csv")
data_2019_07 = pd.read_csv("data/flightlist_20190701_20190731.csv")
data_2019_08 = pd.read_csv("data/flightlist_20190801_20190831.csv")
data_2019_09 = pd.read_csv("data/flightlist_20190901_20190930.csv")
data_2019_10 = pd.read_csv("data/flightlist_20191001_20191031.csv")
data_2019_11 = pd.read_csv("data/flightlist_20191101_20191130.csv")
data_2019_12 = pd.read_csv("data/flightlist_20191201_20191231.csv")
                           
```

```python
frames = [data_2019_01, data_2019_02, data_2019_03, data_2019_04, data_2019_05, data_2019_06, data_2019_07, data_2019_08, data_2019_09, data_2019_10, data_2019_11, data_2019_12]

data_2019_01_12 = pd.concat(frames)

print(data_2019_01_12.shape)
```

```python
data_2019_01_12.head()
```

```python
data_2019_01_12_raw = data_2019_01_12[["number","typecode", "origin", "destination", "day"]].reset_index(drop=True)
```

```python
data_2019_01_06_raw.to_csv('data/data_2020_01_06.original.csv', index=False)
```

```python
data_2019_01_12_filtered = data_2019_01_12_raw.dropna(subset=['origin', 'destination'])
data_2019_01_12_filtered = data_2019_01_12_filtered.reset_index(drop=True)
data_2019_01_12_filtered.shape
```

```python
data_2019_01_12_filtered = data_2019_01_12_filtered[["number","typecode", "origin", "destination", "day"]]
```

```python
data_2019_01_12_filtered.head()
```

```python
data_2019_01_06_filtered.to_csv('data/data_2020_01_06_filtered.csv', index=False)
```

### 2. Combine flight data with airport data

```python
airports = pd.read_csv('data/airports.csv')
```

### 3. Merge flight info & airports 

```python
origin_airports = airports
destination_airports = airports
```

```python
origin_airports.columns = ["origin_ident","origin_type", "origin_name", "origin_latitude_deg", 
                     "origin_longitude_deg", "origin_continent", "origin_iso_country", "origin_iso_region", "origin_municipality" ]
```

```python
origin_airports.shape
```

```python
data_2019_01_12_filtered.shape
```

```python
results = pd.merge(data_2019_01_12_filtered, origin_airports, how="left", left_on = "origin", right_on = "origin_ident")
results
```

```python
destination_airports.columns = ["destination_ident","destination_type", "destination_name", "destination_latitude_deg", 
                                "destination_longitude_deg", "destination_continent", "destination_iso_country", "destination_iso_region", 
                                "destination_municipality" ]

```

```python
results = pd.merge(results, destination_airports, how="left", left_on = "destination", right_on = "destination_ident")
results
```

```python
results.to_csv("data/flight_airport_combination_2019.csv", index=True)
```
